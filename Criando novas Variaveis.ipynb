## Criando Novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Valor bruto soma do valor + condominio + iptu
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Valor do metro quadrado dividindo valor pela area
dados['Valor metro quadarado'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
dados['Valor metro quadarado'] = dados['Valor'] / dados['Area']
#Arredondamento duas casas decimais
dados.round(2).head(10)
```

```python
# Valor do metro quadrado bruto
dados['Valor metro quadarado bruto'] = dados['Valor Bruto'] / dados['Area']
dados.round(2).head(10)
```

```python
# 
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
