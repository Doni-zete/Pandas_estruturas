## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Criar lista de imoveis sem repetição
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',
]
```

```python
residencial
```

```python
dados.head(10)
```

```python
selecao_moveis = dados["Tipo"].isin(residencial)
selecao_moveis
```

```python
dados_residencial = dados[selecao_moveis]
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
len(dados_residencial)
```

```python
len(dados)
```

```python
dados_residencial.index = range(len(dados_residencial))
dados_residencial
```
